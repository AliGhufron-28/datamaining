```python
# Proximity Measure for Nominal Attributes
def d1(i,j):
    P = 2
    Q = 0
    for col in ["Address","Zip"]:
        if dt[col][i] == dt[col][j]:
            Q += 1
    return (P-Q)/P
```

```python
print(d1(1,2))
print(d1(1,3))
print(d1(1,4))
```

```python
# Distance on Numeric Data: Minkowski Distance
def d2(i,j, h):
    result = 0
    for col in ["Price","Area","Room","Lon","Lat"]:
        result = math.pow(abs(dt[col][i] - dt[col][j]), h)
    return result**(1/h)
```

```python
# Manhattan distance
print(d2(1,2,1))
print(d2(1,3,1))
print(d2(1,4,1))
```

```python
# Euclidean distance
print(d2(1,2,2))
print(d2(1,3,2))
print(d2(1,4,2))
```

Disimilarity


```python
# d(1,2)
print(d1(1,2)+d2(1,2,1))
```

```python
# d(1,3)
print(d1(1,3)+d2(1,3,1))
```

```python
# d(1,4)
print(d1(1,4)+d2(1,4,1))
```
