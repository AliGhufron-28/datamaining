## 1. KNN

```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
dt = pd.read_csv("https://raw.githubusercontent.com/AliGhufron-28/datamaining/main/dataR2.csv")
```

```python
dt
```

```python
dt.shape
```

```python
# hapus kolom klasifikasi
X = dt.drop(columns=["Classification"])
X.head()
```

```python
# melihat class kolom Classification
y = dt["Classification"].values
y[0:116]
```

```python
# split dataset into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0, stratify=y)
```

```python
# fitur scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train[:5]
```

```python
from sklearn.neighbors import KNeighborsClassifier
# menentukan nilai k
knn = KNeighborsClassifier(n_neighbors = 5)
# klasifikasi data
knn.fit(X_train,y_train)
```

```python
# menampilkan hasil prediksi dari 5 model
knn.predict(X_test)[0:5]
```

```python
# cek nilai accuracy 
knn.score(X_test, y_test)
```

## 2. Decision Tree

```python
# import library
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn import metrics
from matplotlib import pyplot as plt
```

```python
dt = pd.read_csv("https://raw.githubusercontent.com/AliGhufron-28/datamaining/main/dataR2.csv")
```

```python
dt
```

```python
# melihat klasifikasi dari data
dt.Classification.value_counts()
```

```python
y = dt["Classification"]
X = dt.drop(columns=["Classification"])
clf = tree.DecisionTreeClassifier(criterion="gini")
clf = clf.fit(X, y)
```

```python
fig = plt.figure(figsize=(15,15))
_ = tree.plot_tree(clf, feature_names=list(dt.columns.values)[:9], class_names=list(dt.columns.values)[9] ,filled=True)
```

```python
clf = tree.DecisionTreeClassifier(criterion="gini")
clf = clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
```

```python
print("Accuracy",metrics.accuracy_score(y_test,y_pred))
```

```python
plt.figure(figsize=(15,15))
#create the tree plot
a = tree.plot_tree(clf,
                   rounded = True,
                   filled = True,
                   fontsize=8)
#show the plot
plt.show()
```
